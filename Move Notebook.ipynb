```python
import random
n=4
arr=[[0]*n for i in range(n)]
for i in range(n):
    for j in range(n):
        arr[i][j]=random.randint(0,30)
```

```python
for i in arr:
    print(i)
```

```python
r=31
```

```python
x=arr[0][0]
print(x)
check=0
i=j=0
count=0
print("Rules:")
print("Reach Destination within 50 moves to win.")
print("Numbers and rolled numbers are random and are in the range of 0-30")
print("If you want to skip a turn, enter $")
print("If your current number goes out of range, the game will modulo it by ",r)

while (i!=n-1 and j!=n-1) and count<50:
    p=random.randint(0,10)
    count+=1
    print(count)
    print("Your Current Value is ",x)
    print("You rolled ",p)
    t=input("Enter your operation (+,/,*,-,^,$):")
    if t=='+':
        x=(x+p)%r
    elif t=='-':
        x=abs(x-p)
    elif t=='/':
        try:
            x=x//p
        except:
            print("Division not possible...")
    elif t=='*':
        x=(x*p)%r
    elif t=='^':
        x=(x**p)%r
    elif t=='$':
        x=x
    else:
        print("Invalid Operation... Chance Missed....")
    if i>0 and x==arr[i-1][j]:
        ch=input("Upward Move Possible, Do you want to move upward?(Y/N)")
        if ch=='Y':
            i-=1
            print("Current Position=[",i,",",j,"]")
        else:
            print("You Skipped...")
    elif i<n-1 and x==arr[i+1][j]:
        ch=input("Bottom Move Possible, Do you want to move Bottom?(Y/N)")
        if ch=='Y':
            i+=1
            print("Current Position=[",i,",",j,"]")
        else:
            print("You Skipped...")
    elif j>0 and x==arr[i][j-1]:
        ch=input("Left Move Possible, Do you want to move left?(Y/N)")
        if ch=='Y':
            j-=1
            print("Current Position=[",i,",",j,"]")
        else:
            print("You Skipped...")
    elif j<n-1 and x==arr[i][j+1]:
        ch=input("Right Move Possible, Do you want to move right?(Y/N)")
        if ch=='Y':
            j+=1
            print("Current Position=[",i,",",j,"]")
        else:
            print("You Skipped...")
    if i==n-1 and j==n-1:
        print("CONGRATULATIONS.... YOU WON!!!!!")
        check=1
        break
if check==0:
    print("Sorry you lost... Better lick next time..")
else:
    print("Current Score = ",count)
```
